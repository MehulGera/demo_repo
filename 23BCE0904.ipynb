```python
#Importing relevant libraries and melb_data dataset

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("https://raw.githubusercontent.com/salemprakash/EDA/main/Data/melb_data.csv");
df.head(30)
```

```python
df.shape # tells us the dimension of the dataset in the form of (r,c)
```

**Summary Statistics**

```python
df.describe() # gives the summary statistics of the dataset
```

Checking Dataset for missing values using **isnull()**

```python
df.isnull() #True represents the missing values "NaN" in our Dataset
```

Total number of missing values in dataset

```python
df.isnull().sum().sum()
```

```python
new_df = df.copy()
new_df.isna().sum()
```

```python
#We'll handle each column based on the type of data

new_df = df.copy()
new_df['BuildingArea'] = df['BuildingArea'].fillna(method='bfill') #for BuildingArea we will use the back filling method to fill the missing spaces as data is not continuous

med = df['YearBuilt'].median() #for YearBuilt we will use median to fill the empty values
med2 = df['Car'].median()
new_df['Car'] = df['Car'].fillna(med2) #As the data is discrete and most likely 0,1,2 Median is the  best option to fill the empty cells in this column
series = df['CouncilArea'].mode() # We'll use mode to replace the NaN values to as the values are categorical
mode1 = series[0]


new_df['YearBuilt'] = df['YearBuilt'].fillna(med)
new_df['Car'] = df['Car'].fillna(med2)
new_df['CouncilArea'] = df['CouncilArea'].fillna(mode1)




```

```python
new_df.head(10)
```

```python
new_df.isna().sum().sum()
```

**DATA VISUALIZATION**

```python
#Analyzing the data types in data set
df.dtypes
```

**Correcting Data Types**

```python
new_df['Date'] = pd.to_datetime(new_df['Date'], format='%d/%m/%Y') #Changing the format to date

#below mentioned data types have been converted from float to int64
new_df['Car'] = new_df['Car'].astype('int64')
new_df['YearBuilt'] = new_df['YearBuilt'].astype('int64')
new_df['Propertycount'] = new_df['Propertycount'].astype('int64')

new_df.head(20)
```

**UNIVARIATE ANAYLYSIS**

```python
#1 FacetGrid for Price column
sns.FacetGrid(new_df).map(sns.distplot, "Price").add_legend()
plt.title("Price Distribution 23BCE0904")
plt.xlabel("Price")
```

```python
#2 Bar chart for top property sellers
new_df['SellerG'].value_counts().nlargest(10).plot(kind='bar', figsize=(10,6))
plt.title("Top sellers 23BCE0904")
plt.ylabel('Number of Properties')
plt.xlabel('Seller Group')

```

```python
#Box plot to represent property prices
sns.boxplot(x="Price", data=new_df)
plt.title("23BCE0904")

```

**BIVARIATE ANALYSIS**

```python
# Scatter Plot of Price and Building area
plt.scatter(df["Price"], df["BuildingArea"])
plt.title("Scatter Plot: Price vs Building Area")
plt.xlabel("BuildingArea")
plt.ylabel("Price")
```

```python
# Using box plot to show Price distribution by property type
sns.boxplot(x="Type", y="Price", data=new_df)
plt.title("Box Plot:Price distribution by type")
plt.show()

```

**MULTIVARIATE ANALYSIS**

```python
#1 Pairplot with regression
sns.pairplot(new_df, vars=['Price', 'Rooms', 'BuildingArea', 'Landsize', 'Distance', 'YearBuilt'], kind="reg")
```

```python
#2 Pair plot with hue
sns.set(style="ticks", color_codes=True)
sns.pairplot(new_df, height=2, vars=['Price', 'Rooms', 'BuildingArea'], hue="Type")
plt.show()
```

```python
#3 Calculating correlation coefficient for Price and BuildingArea
from scipy import stats

corr = stats.pearsonr(new_df['Price'], new_df['BuildingArea'])
print("p-value:\t", corr[1])
print("correlation:\t", corr[0])

correlation = new_df.corr(method = 'pearson',numeric_only=True)
print(correlation)

```

```python
# Heatmap

plt.figure(figsize=(10, 8))
sns.heatmap(correlation,xticklabels=correlation.columns, yticklabels=correlation.columns)
plt.title("correlation Heatmap 23BCE0904")
plt.show()
```
